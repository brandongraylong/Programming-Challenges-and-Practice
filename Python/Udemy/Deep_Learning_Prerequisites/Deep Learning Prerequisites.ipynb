```python
# first: eigenvalues, second: eigenvectors
```

```python
Lam, V = np.linalg.eig(A)
```

```python
V[:,0] * Lam[0] == A @ V[:,0]
```

```python
V[:,0] * Lam[0], A @ V[:,0]
```

```python
np.allclose(V[:,0] * Lam[0], A @ V[:,0])
```

```python
np.allclose(V @ np.diag(Lam), A @ V)
```
